```python

# s = S()
s = '1'
_len = len(s)

if _len == 1:
    print(0)
elif _len % 2 == 0:
    # 桁数が偶数の場合
    # 奇数の時も多分同じロジックになる？
    a = int(_len / 2)
    b = int(s[:a])
    c = int(s[a:])
    if b > c:
        b -= 1
    print(b)
else:
    # 奇数の時
    a = int(_len / 2)
    s = "9" * a * 2
    # print(s, a)
    b = int(s[:a])
    c = int(s[a:])
    if b > c:
        b -= 1
    print(b)
```
